# 13. Using the user defined function convert the ‘rasied_amount_used’ column into a categorical column as follow

# a) If the amount is less than 5 Millions then recode as TypeA.
# b) If the amount is greater than 5 Millions and less than 7 Millions then recode as TypeB.
# c) If the amount is greater than 7 Millions then recode as TypeC

```python
# created user defined function to categorize "raised_amount_usd" column into 3 different types as given in question statement.
def new_category(x):
    s=[]
    if x<5000000:
        s.append('typeA')
    if x>5000000 and x<7000000:
        s.append('typeB')
    if x>7000000:
        s.append('typeC')
    se=pd.Series(s)
    se_val=se.values
    return(se_val)       
```

```python
# Applying lambda function to new column to list down type categories
master_dataframe['new_1']=master_dataframe['raised_amount_usd'].apply(lambda x: new_category(x))
```
